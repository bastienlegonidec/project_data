```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import pandas as pd
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('result.csv')

classe_col = df['bug type']

columns = df.columns.tolist()
columns.remove('species')
columns.remove('bug type')

classes = np.unique(classe_col)
print(classes)
```

```python
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df[columns])
```

```python
fig, ax = plt.subplots()
for classe in classes:
    ax.scatter(df_pca[classe_col == classe, 0], df_pca[classe_col == classe, 1], label=classe)

ax.legend()
ax.set_xlabel("PC1 {0}% of explained variance".format(round(pca.explained_variance_ratio_[0]*100,2), fontsize = 12))
ax.set_ylabel("PC2 {0}% of explained variance".format(round(pca.explained_variance_ratio_[1]*100,2), fontsize = 12))
ax.set_title('PCA space', fontsize = 12)
```
